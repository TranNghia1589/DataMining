```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

data = pd.read_csv(r'D:\DataWarehouse and DataMining\DataMining\BaoCaoNhom\file.csv')
data
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Tạo bảng pivot
basket = pd.pivot_table(data=data,
                       index='Transaction_ID',
                       columns='Product_Description',
                       values='Quantity',
                       aggfunc='sum',
                       fill_value=0)

# Chuyển đổi thành dạng binary (0,1)
basket_sets = (basket > 0).astype(int)
```

```python
basket_sets.head()
```

```python
# Tìm các tập phổ biến với min_support
frequent_itemsets = apriori(basket_sets, 
                          min_support=0.01,  # Xuất hiện ít nhất 3%
                          use_colnames=True)
```

```python
frequent_itemsets.head(100)
```

```python
# Tạo rules với min_confidence
rules = association_rules(frequent_itemsets,
                        metric="lift",
                        min_threshold=0.6,
                        num_itemsets=len(frequent_itemsets))  # Độ tin cậy tối thiểu 60%

# Sắp xếp theo lift
rules = rules.sort_values('lift', ascending=False)
```

```python
rules.head(100)
```

```python
# Lọc rules theo metrics
rules[
    (rules['confidence'] >= 0.2) &  # Độ tin cậy cao
    (rules['lift'] > 1.5)        # Lift > 1 chỉ ra mối quan hệ tích cực
]
```
